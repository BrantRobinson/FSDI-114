# Problem 1
## Find the total of numbers from a collection of numbers that divided by 2 the module id 0

Examples:
Given [3,4,10,15,999,2] then the output should be  **total = 3**
Given [2, 4, 6, 10, 3.7] then the output should be  **total = 4**

### hint:
- '%' is the remainder from division
- 2 % 2 = 0

collection of numbers is [3,5,27,20,6,3,5,60]


```python
number_list = [3,5,27,20,6,3,5,60]

def count_evens(numbers):
    count = 0
    for number in numbers:
        remainder = number % 2
        if remainder == 0:
            count += 1
    return count

count = count_evens(number_list)
print(f"total = {count}")
```

```python
number_list = [3,5,27,20,6,3,5,60]

def count_evens(numbers):
    count = 0
    evens = []
    for number in numbers:
        if number % 2 == 0:
            count += 1
            evens.append(number)
    return count, evens

count, evens = count_evens(number_list)
print(f"total = {count}, list: {evens}")
```

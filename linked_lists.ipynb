```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None  

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return    
        
        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node


    def print_list(self):
        current = self.head
        if not current:
            print("List is empty")
            return
        while current:
            print(current.value)
            current = current.next

    def count_nodes(self):
        nodes = 0
        current = self.head
        while current:
            nodes += 1
            current = current.next
        return nodes
    
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    def remove_node(self, value):
        current = self.head
        if not current:
            return
        if current.value == value:
            self.head = current.next
        while current.next:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next
          

new_list = LinkedList()
for i in range(10, 40, 10):
    new_list.append(i)
print("Original List")
new_list.print_list()
print("Node Count")
print(new_list.count_nodes())
print("Remove 20")
new_list.remove_node(20)
new_list.print_list()
print("Prepend 99")
new_list.prepend(99)
new_list.print_list()


```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None  

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return    
        
        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node


    def print_list(self):
        current = self.head
        if not current:
            print("List is empty")
            return
        while current:
            print(current.value)
            current = current.next

    def count_nodes(self):
        nodes = 0
        current = self.head
        while current:
            nodes += 1
            current = current.next
        return nodes
    
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    def remove_node(self, value):
        current = self.head
        if not current:
            return
        if current.value == value:
            self.head = current.next
        while current.next:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next
          

new_list = LinkedList()
for i in range(10, 40, 10):
    new_list.append(i)
print("Original List")
new_list.print_list()
print("Node Count")
print(new_list.count_nodes())
print("Remove 20")
new_list.remove_node(20)
new_list.print_list()
print("Prepend 99")
new_list.prepend(99)
new_list.print_list()


```

```python
# Double linked list
class DLLNode:
    def __init__(self, value):
        self.value = value
        self.next = None  
        self.prev = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = DLLNode(value)

        if not self.head:
            self.head = self.tail = new_node
            return

        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node


    def print_fwd(self):
        current = self.head
        if not current:
            print("List is empty")
            return
        while current:
            if current.value is not None:
                print(current.value)
            current = current.next

    def print_rev(self):
        current = self.tail
        if not current:
            print("List is empty")

        while current:
            print(current.value)
            current = current.prev

    def prepend(self, value):
        #create the new node
        new_dll_node = DLLNode(value)

        #if this is the first node - just do what the append does and set the new one to head and tail
        if not self.head:
            self.head = self.tail = new_dll_node
            return

        #grab the current head
        current = self.head
        #replace the current head
        self.head = new_dll_node
        #old current is now the second node so...
        self.head.next = current
        #need to change prev for the old head, next should stay the same
        current.prev = new_dll_node

    def remove(self,value):
        # if there is nothing in the list just exit the function
        if not self.head:
            print("The list is empty")
            return
        current = self.head
        current_tail = self.tail

        ## check to see if there is only one value
        if self.head == self.tail:
            if self.head.value == value:
                self.head = None
                self.tail = None
            else:
                print("Value not found")
            return


        #check to see if the value they are looking for is the head 
        if current.value == value:
            #if this is true then you have to make a new head which would be the next value
            self.head = current.next
            #set the new head previous to None
            if self.head:
                self.head.prev = None
            return
        
        #check to see if the value they are looking for is the tail 
        if current_tail.value == value:
            #if this is true then I have to make a new tail which would be the previous value
            self.tail = current_tail.prev
            #set the new tail next to none
            if self.tail:
                self.tail.next = None 
            return
            
        #if it's not the head then try to find it
        while current:
            if current.value == value:
                # found it so now I need to set the one before it to skip the current
                previous = current.prev
                previous.next = current.next
                # and set the one after it to skip it backwards
                current.next.prev = current.prev
                return
            current = current.next
                
        print("value not found")
            


    def remove_node(self, value):
        current = self.head
        if not current:
            return
        if current.value == value:
            self.head = current.next
        while current.next:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next

new_dll = DoubleLinkedList()
for i in range(5):
    new_dll.append(i)
    
print("--------Print Forward---------")
print(new_dll.print_fwd())
print("--------Print Reverse---------")
print(new_dll.print_rev())

print("------after prepend 99------")
new_dll.prepend(99)
print(new_dll.print_fwd())

print("------after removal 3------")
new_dll.remove(3)
print(new_dll.print_fwd())
        

        
```

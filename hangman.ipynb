```python
import random

fruits = ["apple", "banana", "orange", "grape", "mango"]
already_guessed = []
found_letters_index = []
lives = 6

def pick_random_word():
    random_number = random.randint(0, len(fruits)-1)
    print(random_number)

    random_word = fruits[random_number]
    print(random_word)

    return random_word

def guess_a_letter(random_word):
        guess = input("Guess a letter")
        if len(guess) != 1:
            print("you can only guess 1 letter")
        else:
            already_guessed.append(guess)
            found_index = 0
            for letter in random_word:
                if letter == guess:
                    found_letters_index.append(found_index)
                found_index += 1

        if found_letters_index:
            print (F"found letters index: {found_letters_index}")
        else:
            print (f"you didn't find a letter")

random_word = pick_random_word()

while lives > 0:
    guess_a_letter(random_word)


    
        
```
